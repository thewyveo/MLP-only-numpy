```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the column names based on the attribute information
columns = [
    'class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
    'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
    'stalk-shape', 'stalk-surface-above-ring', 'stalk-surface-below-ring',
    'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type',
    'veil-color', 'ring-number', 'ring-type', 'spore-print-color',
    'population', 'habitat'
]

# Load the data
data = pd.read_csv('processed-mushroom.data', header=None, names=columns)

# Display the first few rows of the dataset
print(data.head())
```

```python
# Count the occurrences of each odor type
odor_distribution = data['odor'].value_counts()

# Plot the distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=odor_distribution.index, y=odor_distribution.values, palette='viridis')
plt.title('Distribution of Odor Types')
plt.xlabel('Odor')
plt.ylabel('Count')
plt.show()
```

```python
# Create a cross-tabulation of gill-size and class
gill_size_class = pd.crosstab(data['gill-size'], data['class'])

# Plot the correlation
gill_size_class.plot(kind='bar', stacked=True, figsize=(8, 6), colormap='viridis')
plt.title('Correlation between Gill Size and Class')
plt.xlabel('Gill Size')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data (assuming it's already loaded as `data`)
# If not, load it as shown in the previous example.

# Select the relevant columns for the plot
subset_data = data[['class', 'odor', 'spore-print-color']]

# Create a grouped bar plot
plt.figure(figsize=(12, 8))
sns.countplot(data=subset_data, x='odor', hue='spore-print-color', palette='viridis', 
              order=sorted(subset_data['odor'].unique()), 
              hue_order=sorted(subset_data['spore-print-color'].unique()))

# Add labels and title
plt.title('Distribution of Odor and Spore Print Color by Mushroom Class', fontsize=16)
plt.xlabel('Odor', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Spore Print Color', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data (assuming it's already loaded as `data`)
# If not, load it as shown in the previous example.

# Select the relevant columns for the plot
subset_data = data[['class', 'odor']]

# Create a grouped bar plot
plt.figure(figsize=(10, 6))
sns.countplot(data=subset_data, x='odor', hue='class', palette='viridis', 
              order=sorted(subset_data['odor'].unique()))

# Add labels and title
plt.title('Distribution of Odor by Mushroom Class (Edible vs. Poisonous)', fontsize=16)
plt.xlabel('Odor', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.legend(title='Class', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
